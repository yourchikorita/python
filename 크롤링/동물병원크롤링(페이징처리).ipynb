## a에 class가 있을때


```
from bs4 import BeautifulSoup
import requests

maximum = 0
page = 1

URL = 'http://www.animal.go.kr/portal_rnl/overall_info/hospital.jsp?s_date=&e_date=&s_upr_cd=&s_org_cd=&s_up_kind_cd=&s_kind_cd=&s_name=&s_shelter_cd=&s_wrk_cd=&s_state=&s_state_hidden=&pagecnt=1'
response = requests.get(URL)
source = response.text
soup = BeautifulSoup(source, 'html.parser')
```

```
while 1:
	page_list = soup.findAll("a", {"class": "NP=r:" + str(page)})
	if not page_list:
		maximum = page - 1
		break
	page = page + 1
print("총 " + str(maximum) + " 개의 페이지가 확인 됬습니다.")
```

```
whole_source = ""
for page_number in range(1, maximum+1):
	URL = 'http://land.naver.com/news/field.nhn?page=' + str(page_number)
	response = requests.get(URL)
	whole_source = whole_source + response.text
soup = BeautifulSoup(whole_source, 'html.parser')
find_title = soup.select("#content > div.section_headline > ul > li > dl > dt > a")

for title in find_title:
	print(title.text)
```

## 두번째 크롤링 동물병원

```
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
base_url = 'http://www.animal.go.kr/portal_rnl/overall_info/hospital.jsp?s_date=&e_date=&s_upr_cd=&s_org_cd=&s_up_kind_cd=&s_kind_cd=&s_name=&s_shelter_cd=&s_wrk_cd=&s_state=&s_state_hidden=&pagecnt={}'
hospiNames_list=[]
hospiAddrs_list=[]
hospiPhone_list=[]

for n in range (478) :
    print(n+1,'번째 page--------------------------------------------------------')
    url = base_url.format(n+1)
    webpage = urlopen(url)
    source = BeautifulSoup(webpage, 'html5lib')
    hospiNames = source.find_all('td', {'class':'hospiName'})
    hospiAddrs= source.find_all('td', {'class':'hospiAddr'})
    hospiPhones= source.find_all('td', {'class':'hospiPhone'})
    # hospiNames_list=[]

    for hospiName in hospiNames :
      hospiNames_list.append(hospiName.get_text().strip())
    for hospiAddr in hospiAddrs :
      hospiAddrs_list.append(hospiAddr.get_text().strip())  
    for hospiPhone in hospiPhones :
      hospiPhone_list.append(hospiPhone.get_text().strip())  

print('hospiNames_list',len(hospiNames_list))
print(hospiNames_list)
print('hospiAddrs_list',len(hospiAddrs_list))
print(hospiAddrs_list)
print('hospiPhone_list',len(hospiPhone_list))
print(hospiPhone_list)



# 데이타프레임만들기
df=pd.DataFrame(data=dict(병원명=hospiNames_list,소재지=hospiAddrs_list,전화번호=hospiPhone_list))

df.to_csv('pet_hospital.csv')
```
